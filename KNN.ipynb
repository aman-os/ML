## KNeighborsClassifier

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris=pd.read_csv(r"C:\Users\amanc\Downloads\iris.csv")
iris
```

```python
iris.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
iris['species']=LabelEncoder().fit_transform(iris['species'])
```

```python
x=iris.drop(['species'],axis=1)
```

```python
y=iris['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
model=KNeighborsClassifier()
model.fit(x_train,y_train)
```

```python
train_pred=model.predict(x_train)
```

```python
test_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(train_pred,y_train)
```

```python
accuracy_score(test_pred,y_test)
```

```python
x=diam.drop(["price"],axis=1)
```

```python
y=diam[["price"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
model=KNeighborsRegressor(metric="euclidean")
model.fit(x_train,y_train)
```

```python
train_pred=model.predict(x_train)
test_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(train_pred,y_train)
```

```python
r2_score(test_pred,y_test)
```
